```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("../data/cleaned_car_data.csv")

df.head()
```

Top Car Companies

```python
top_companies = df["company"].value_counts().head(10)

fig = px.bar(
    top_companies,
    title="Top 10 Car Companies"
)

fig.show()
```

```python
fig = px.pie(
    df,
    names="fuel_type",
    title="Fuel Type Distribution"
)

fig.show()
```

```python
fig = px.histogram(
    df,
    x="Price",
    title="Price Distribution"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="year",
    y="Price",
    color="fuel_type",
    hover_data=["company"]
)

fig.show()
```
